First, I use rectangle images just including each building.

### Define functions

Before defining a function to clean and save image data, I define a function to transform coordinates.

```python
def transform_coordinates(coordinates_list, transform):
    
    transformed_coordinates_list = []
    for coordinate in coordinates_list:
        transformed_coordinate = list(transform(coordinate[0], coordinate[1]))
        transformed_coordinates_list.append(transformed_coordinate)

    return transformed_coordinates_list
```

```python
def data_clean_save(country, place, mode = 'resize'):

  path = '/content/drive/My Drive/stac/'

  # Training geojson file
  train_geojson_file = 'train-' + place + '.geojson'
  train_gpd = gpd.read_file(os.path.join(path, country, place, train_geojson_file))

  # Test geojson file
  test_geojson_file = 'test-' + place + '.geojson'
  test_gpd = gpd.read_file(os.path.join(path, country, place, test_geojson_file))

  # Tif file
  tif_file = place + '_ortho-cog.tif'
  raster_ds = rasterio.open(os.path.join(path, country, place ,tif_file))
  proj = Proj(init = raster_ds.crs.data['init'])

  # Clean GPD datasets =========

  # Convert Multypolygons to single Polygons and choose the biggest ones ----------
  # Training data
  train_mp_obs = train_gpd.geometry.apply(lambda x: x.type == 'MultiPolygon')
  train_gpd_mp = train_gpd[train_mp_obs]

  train_polygon_list = []
  for i in train_gpd_mp.index:

    ss = train_gpd.geometry[i]
    polygon_area = 0
    for index, polygon in enumerate(ss):
      if polygon.area > polygon_area:
        single_polygon = polygon
        polygon_area = polygon.area

    train_polygon_list.append(single_polygon)

  train_gpd_clean = copy(train_gpd)
  train_gpd_clean.geometry[train_mp_obs] = train_polygon_list

  # Test data
  test_mp_obs = test_gpd.geometry.apply(lambda x: x.type == 'MultiPolygon')
  test_gpd_mp = test_gpd[test_mp_obs]

  test_polygon_list = []
  for i in test_gpd_mp.index:

    ss = test_gpd.geometry[i]
    polygon_area = 0
    for index, polygon in enumerate(ss):
      if polygon.area > polygon_area:
        single_polygon = polygon
        polygon_area = polygon.area

    test_polygon_list.append(single_polygon)

  test_gpd_clean = copy(test_gpd)
  test_gpd_clean.geometry[test_mp_obs] = test_polygon_list

  # Get centroids ===============

  # Training data
  train_gpd_clean['centroid_x'] = train_gpd_clean.geometry.apply(lambda x: x.exterior.centroid.xy[0][0])
  train_gpd_clean['centroid_y'] = train_gpd_clean.geometry.apply(lambda x: x.exterior.centroid.xy[1][0])
  train_gpd_clean_coords = train_gpd_clean[['centroid_x', 'centroid_y']]

  # Test data
  test_gpd_clean['centroid_x'] = test_gpd_clean.geometry.apply(lambda x: x.exterior.centroid.xy[0][0])
  test_gpd_clean['centroid_y'] = test_gpd_clean.geometry.apply(lambda x: x.exterior.centroid.xy[1][0])
  test_gpd_clean_coords = test_gpd_clean[['centroid_x', 'centroid_y']]

  # Concatenate training and test dataset coordinates
  centroid_coords = pd.concat([train_gpd_clean_coords, test_gpd_clean_coords])

  # Extract image data and save as numpy arrays ==================
  # For training data

  label_series = train_gpd_clean.roof_material

  log_building_area_list = []
  building_vertices_list = []
  building_width_list = []
  building_height_list = []
  R_mean, G_mean, B_mean = [], [], []
  R_std, G_std, B_std = [], [], []
  H_mean, S_mean, V_mean = [], [], []
  H_std, S_std, V_std = [], [], []

  for i in range(len(train_gpd_clean)):

    building_id = train_gpd_clean.id[i]
    building = train_gpd_clean.geometry[i]
    lon, lat = building.exterior.coords.xy
    coord_list = [[np.min(lon), np.min(lat)],
                 [np.min(lon), np.max(lat)],
                 [np.max(lon), np.max(lat)],
                 [np.max(lon), np.min(lat)]]
    transformed_coord_list = transform_coordinates(coord_list, proj)
    dict_transformed_coord = {'coordinates': [transformed_coord_list], 'type': 'Polygon'}
    file_name_sq = building_id + '.jpg'
    if mode == 'resize':
      out_image, out_transform = mask(raster_ds, [dict_transformed_coord], crop=True)      
      hsv_out_image = (rasterio.plot.reshape_as_image(out_image)[:,:,:3] / 255.).astype(np.float32)
      hsv_out_image = cv2.cvtColor(hsv_out_image, cv2.COLOR_RGB2HSV)
      hsv_out_image[:,:,0] = hsv_out_image[:,:,0] / 360
      plt.imsave(os.path.join(path, 'image_resize/train', file_name_sq), rasterio.plot.reshape_as_image(out_image)[:,:,:3])
      plt.imsave(os.path.join(path, 'image_resize_hsv/train', file_name_sq), hsv_out_image)

    elif mode == 'mask':
      polygon_coord_list = [[lon[i], lat[i]] for i in range(len(lon))]
      transformed_polygon_coord_list = transform_coordinates(polygon_coord_list, proj)
      dict_transformed_polygon_coord = {'coordinates': [transformed_polygon_coord_list], 'type': 'Polygon'}
      out_image, out_transform = mask(raster_ds, [dict_transformed_polygon_coord], crop=True)
      file_name = building_id + '.jpg'
      plt.imsave(os.path.join(path, 'image_mask/train', file_name), rasterio.plot.reshape_as_image(out_image)[:,:,:3])
      hsv_out_image = (rasterio.plot.reshape_as_image(out_image)[:,:,:3] / 255.).astype(np.float32)
      hsv_out_image = cv2.cvtColor(hsv_out_image, cv2.COLOR_RGB2HSV)
      hsv_out_image[:,:,0] = hsv_out_image[:,:,0] / 360
      plt.imsave(os.path.join(path, 'image_mask_hsv/train', file_name_sq), hsv_out_image)

    newimg = cv2.resize(rasterio.plot.reshape_as_image(out_image), (244, 244))
    newimg_hsv = cv2.resize(hsv_out_image, (244, 244))
    R_mean.append(newimg[:, :, 1].mean())
    G_mean.append(newimg[:, :, 2].mean())
    B_mean.append(newimg[:, :, 3].mean())
    R_std.append(newimg[:, :, 1].std())
    G_std.append(newimg[:, :, 2].std())
    B_std.append(newimg[:, :, 3].std())
    H_mean.append(newimg_hsv[:, :, 0].mean())
    S_mean.append(newimg_hsv[:, :, 1].mean())
    V_mean.append(newimg_hsv[:, :, 2].mean())
    H_std.append(newimg_hsv[:, :, 0].std())
    S_std.append(newimg_hsv[:, :, 1].std())
    V_std.append(newimg_hsv[:, :, 2].std())

    log_building_area_list.append(np.log(building.area))
    building_vertices_list.append(len(lon))
    building_width_list.append(geopy.distance.distance((np.max(lat), np.max(lon)), (np.max(lat), np.min(lon))).meters)
    building_height_list.append(geopy.distance.distance((np.max(lat), np.max(lon)), (np.min(lat), np.max(lon))).meters)

  id_series = train_gpd_clean.id
  train_pd = pd.DataFrame({'id': id_series, 'label': label_series})
  train_pd = pd.concat([train_pd, train_gpd_clean_coords], axis = 1)
  train_pd = pd.concat([train_pd, 
                        pd.DataFrame({'log_building_area': log_building_area_list, 
                                      'building_vertices': building_vertices_list,
                                      'building_width': building_width_list,
                                      'building_height': building_height_list,
                                      'place': place,
                                      'R_mean': R_mean, 'G_mean': G_mean, 'B_mean': B_mean,
                                      'R_std': R_std, 'G_std': G_std, 'B_std': B_std,
                                      'H_mean': H_mean, 'S_mean': S_mean, 'V_mean': V_mean,
                                      'H_std': H_std, 'S_std': S_std, 'V_std': V_std
                                      })], 
                       axis = 1)

  # For test data

  log_building_area_list = []
  building_vertices_list = []
  building_width_list = []
  building_height_list = []
  R_mean, G_mean, B_mean = [], [], []
  R_std, G_std, B_std = [], [], []
  H_mean, S_mean, V_mean = [], [], []
  H_std, S_std, V_std = [], [], []

  for i in range(len(test_gpd_clean)):

    building_id = test_gpd_clean.id[i]
    building = test_gpd_clean.geometry[i]
    lon, lat = building.exterior.coords.xy
    coord_list = [[np.min(lon), np.min(lat)],
                 [np.min(lon), np.max(lat)],
                 [np.max(lon), np.max(lat)],
                 [np.max(lon), np.min(lat)]]
    transformed_coord_list = transform_coordinates(coord_list, proj)
    dict_transformed_coord = {'coordinates': [transformed_coord_list], 'type': 'Polygon'}
    file_name_sq = building_id + '.jpg'  
    if mode == 'resize':
      out_image, out_transform = mask(raster_ds, [dict_transformed_coord], crop=True)
      # plt.imsave(os.path.join(path, 'image_resize/test', file_name_sq), rasterio.plot.reshape_as_image(out_image)[:,:,:3])
      hsv_out_image = (rasterio.plot.reshape_as_image(out_image)[:,:,:3] / 255.).astype(np.float32)
      hsv_out_image = cv2.cvtColor(hsv_out_image, cv2.COLOR_RGB2HSV)
      hsv_out_image[:,:,0] = hsv_out_image[:,:,0] / 360
      plt.imsave(os.path.join(path, 'image_resize_hsv/test', file_name_sq), hsv_out_image)

    elif mode == 'mask':
      polygon_coord_list = [[lon[i], lat[i]] for i in range(len(lon))]
      transformed_polygon_coord_list = transform_coordinates(polygon_coord_list, proj)
      dict_transformed_polygon_coord = {'coordinates': [transformed_polygon_coord_list], 'type': 'Polygon'}
      out_image, out_transform = mask(raster_ds, [dict_transformed_polygon_coord], crop=True)
      file_name = building_id + '.jpg'
      plt.imsave(os.path.join(path, 'image_mask/test', file_name), rasterio.plot.reshape_as_image(out_image)[:,:,:3])
      hsv_out_image = (rasterio.plot.reshape_as_image(out_image)[:,:,:3] / 255.).astype(np.float32)
      hsv_out_image = cv2.cvtColor(hsv_out_image, cv2.COLOR_RGB2HSV)
      hsv_out_image[:,:,0] = hsv_out_image[:,:,0] / 360
      plt.imsave(os.path.join(path, 'image_mask_hsv/test', file_name_sq), hsv_out_image)

    newimg = cv2.resize(rasterio.plot.reshape_as_image(out_image), (244, 244))
    newimg_hsv = cv2.resize(hsv_out_image, (244, 244))
    R_mean.append(newimg[:, :, 1].mean())
    G_mean.append(newimg[:, :, 2].mean())
    B_mean.append(newimg[:, :, 3].mean())
    R_std.append(newimg[:, :, 1].std())
    G_std.append(newimg[:, :, 2].std())
    B_std.append(newimg[:, :, 3].std())
    H_mean.append(newimg_hsv[:, :, 0].mean())
    S_mean.append(newimg_hsv[:, :, 1].mean())
    V_mean.append(newimg_hsv[:, :, 2].mean())
    H_std.append(newimg_hsv[:, :, 0].std())
    S_std.append(newimg_hsv[:, :, 1].std())
    V_std.append(newimg_hsv[:, :, 2].std())

    log_building_area_list.append(np.log(building.area))
    building_vertices_list.append(len(lon))
    building_width_list.append(geopy.distance.distance((np.max(lat), np.max(lon)), (np.max(lat), np.min(lon))).meters)
    building_height_list.append(geopy.distance.distance((np.max(lat), np.max(lon)), (np.min(lat), np.max(lon))).meters)
    
  id_series = test_gpd_clean.id
  test_pd = pd.DataFrame({'id': id_series})
  test_pd = pd.concat([test_pd, test_gpd_clean_coords], axis = 1)
  test_pd = pd.concat([test_pd, 
                        pd.DataFrame({'log_building_area': log_building_area_list, 
                                      'building_vertices': building_vertices_list,
                                      'building_width': building_width_list,
                                      'building_height': building_height_list,
                                      'place': place,
                                      'R_mean': R_mean, 'G_mean': G_mean, 'B_mean': B_mean,
                                      'R_std': R_std, 'G_std': G_std, 'B_std': B_std,
                                      'H_mean': H_mean, 'S_mean': S_mean, 'V_mean': V_mean,
                                      'H_std': H_std, 'S_std': S_std, 'V_std': V_std
                                      })], 
                       axis = 1)

  return train_pd, test_pd

```

## Calculate variables related to distances between buildings

Here, I define a function to calculate distances from buildings to neighboring buildings and calculate summary statistics related the distances.

```python
def data_clean_distance_save(country, place):

  path = '/content/drive/My Drive/stac/'

  # Training geojson file
  train_geojson_file = 'train-' + place + '.geojson'
  train_gpd = gpd.read_file(os.path.join(path, country, place, train_geojson_file))

  # Test geojson file
  test_geojson_file = 'test-' + place + '.geojson'
  test_gpd = gpd.read_file(os.path.join(path, country, place, test_geojson_file))

  # Tif file
  tif_file = place + '_ortho-cog.tif'
  raster_ds = rasterio.open(os.path.join(path, country, place ,tif_file))
  proj = Proj(init = raster_ds.crs.data['init'])

  # Clean GPD datasets =========

  # Convert Multypolygons to single Polygons and choose the biggest ones ----------
  # Training data
  train_mp_obs = train_gpd.geometry.apply(lambda x: x.type == 'MultiPolygon')
  train_gpd_mp = train_gpd[train_mp_obs]

  train_polygon_list = []
  for i in train_gpd_mp.index:

    ss = train_gpd.geometry[i]
    polygon_area = 0
    for index, polygon in enumerate(ss):
      if polygon.area > polygon_area:
        single_polygon = polygon
        polygon_area = polygon.area

    train_polygon_list.append(single_polygon)

  train_gpd_clean = copy(train_gpd)
  train_gpd_clean.geometry[train_mp_obs] = train_polygon_list

  # Test data
  test_mp_obs = test_gpd.geometry.apply(lambda x: x.type == 'MultiPolygon')
  test_gpd_mp = test_gpd[test_mp_obs]

  test_polygon_list = []
  for i in test_gpd_mp.index:

    ss = test_gpd.geometry[i]
    polygon_area = 0
    for index, polygon in enumerate(ss):
      if polygon.area > polygon_area:
        single_polygon = polygon
        polygon_area = polygon.area

    test_polygon_list.append(single_polygon)

  test_gpd_clean = copy(test_gpd)
  test_gpd_clean.geometry[test_mp_obs] = test_polygon_list

  # Get centroids ===============

  # Training data
  train_gpd_clean['centroid_x'] = train_gpd_clean.geometry.apply(lambda x: x.exterior.centroid.xy[0][0])
  train_gpd_clean['centroid_y'] = train_gpd_clean.geometry.apply(lambda x: x.exterior.centroid.xy[1][0])
  train_gpd_clean_coords = train_gpd_clean[['centroid_x', 'centroid_y']]

  # Test data
  test_gpd_clean['centroid_x'] = test_gpd_clean.geometry.apply(lambda x: x.exterior.centroid.xy[0][0])
  test_gpd_clean['centroid_y'] = test_gpd_clean.geometry.apply(lambda x: x.exterior.centroid.xy[1][0])
  test_gpd_clean_coords = test_gpd_clean[['centroid_x', 'centroid_y']]

  # Concatenate training and test dataset coordinates
  centroid_coords = pd.concat([train_gpd_clean_coords, test_gpd_clean_coords])

  # Extract image data and save as numpy arrays ==================
  # For training data

  label_series = train_gpd_clean.roof_material

  distance_10_all_list, distance_20_all_list, distance_50_all_list = [], [], []
  distance_10_train_list, distance_20_train_list, distance_50_train_list = [], [], []
  concrete_10_train_list, healthy_10_train_list, incomplete_10_train_list, irregular_10_train_list, other_10_train_list = [], [], [], [], []
  concrete_20_train_list, healthy_20_train_list, incomplete_20_train_list, irregular_20_train_list, other_20_train_list = [], [], [], [], []
  concrete_50_train_list, healthy_50_train_list, incomplete_50_train_list, irregular_50_train_list, other_50_train_list = [], [], [], [], []

  for i in range(len(train_gpd_clean)):

    geod = pyproj.Geod(ellps='WGS84')
    def calculate_distance(x, y):
      return geod.inv(train_gpd_clean_coords.iloc[i,0], train_gpd_clean_coords.iloc[i,1], x, y)[2]
    vec_calculate_distance = np.vectorize(calculate_distance)
    distance_from_other_building_train = vec_calculate_distance(train_gpd_clean_coords.centroid_x, train_gpd_clean_coords.centroid_y)
    distance_from_other_building_all = vec_calculate_distance(centroid_coords.centroid_x, centroid_coords.centroid_y)
    
    distance_10_all = np.sum((distance_from_other_building_all < 10) & (distance_from_other_building_all > 0))
    distance_20_all = np.sum((distance_from_other_building_all < 20) & (distance_from_other_building_all > 0))
    distance_50_all = np.sum((distance_from_other_building_all < 50) & (distance_from_other_building_all > 0))

    distance_10_train = np.sum((distance_from_other_building_train < 10) & (distance_from_other_building_train > 0))
    distance_20_train = np.sum((distance_from_other_building_train < 20) & (distance_from_other_building_train > 0))
    distance_50_train = np.sum((distance_from_other_building_train < 50) & (distance_from_other_building_train > 0))

    concrete_10_train   = (np.sum(label_series[(distance_from_other_building_train < 10) & (distance_from_other_building_train > 0)] == 'concrete_cement') + 1) / (distance_10_train + 5)
    healthy_10_train    = (np.sum(label_series[(distance_from_other_building_train < 10) & (distance_from_other_building_train > 0)] == 'healthy_metal') + 1) / (distance_10_train + 5)
    incomplete_10_train = (np.sum(label_series[(distance_from_other_building_train < 10) & (distance_from_other_building_train > 0)] == 'incomplete') + 1) / (distance_10_train + 5)
    irregular_10_train  = (np.sum(label_series[(distance_from_other_building_train < 10) & (distance_from_other_building_train > 0)] == 'irregular_metal') + 1) / (distance_10_train + 5)
    other_10_train      = (np.sum(label_series[(distance_from_other_building_train < 10) & (distance_from_other_building_train > 0)] == 'other') + 1) / (distance_10_train + 5)

    concrete_20_train   = (np.sum(label_series[(distance_from_other_building_train < 20) & (distance_from_other_building_train > 0)] == 'concrete_cement') + 1) / (distance_20_train + 5)
    healthy_20_train    = (np.sum(label_series[(distance_from_other_building_train < 20) & (distance_from_other_building_train > 0)] == 'healthy_metal') + 1) / (distance_20_train + 5)
    incomplete_20_train = (np.sum(label_series[(distance_from_other_building_train < 20) & (distance_from_other_building_train > 0)] == 'incomplete') + 1) / (distance_20_train + 5)
    irregular_20_train  = (np.sum(label_series[(distance_from_other_building_train < 20) & (distance_from_other_building_train > 0)] == 'irregular_metal') + 1) / (distance_20_train + 5)
    other_20_train      = (np.sum(label_series[(distance_from_other_building_train < 20) & (distance_from_other_building_train > 0)] == 'other') + 1) / (distance_20_train + 5)

    concrete_50_train   = (np.sum(label_series[(distance_from_other_building_train < 50) & (distance_from_other_building_train > 0)] == 'concrete_cement') + 1) / (distance_50_train + 5)
    healthy_50_train    = (np.sum(label_series[(distance_from_other_building_train < 50) & (distance_from_other_building_train > 0)] == 'healthy_metal') + 1) / (distance_50_train + 5)
    incomplete_50_train = (np.sum(label_series[(distance_from_other_building_train < 50) & (distance_from_other_building_train > 0)] == 'incomplete') + 1) / (distance_50_train + 5)
    irregular_50_train  = (np.sum(label_series[(distance_from_other_building_train < 50) & (distance_from_other_building_train > 0)] == 'irregular_metal') + 1) / (distance_50_train + 5)
    other_50_train      = (np.sum(label_series[(distance_from_other_building_train < 50) & (distance_from_other_building_train > 0)] == 'other') + 1) / (distance_50_train + 5)

    distance_10_all_list.append(distance_10_all)
    distance_20_all_list.append(distance_20_all)
    distance_50_all_list.append(distance_50_all)
    distance_10_train_list.append(distance_10_train)
    distance_20_train_list.append(distance_20_train)
    distance_50_train_list.append(distance_50_train)

    concrete_10_train_list.append(concrete_10_train)
    healthy_10_train_list.append(healthy_10_train)
    incomplete_10_train_list.append(incomplete_10_train)
    irregular_10_train_list.append(irregular_10_train)
    other_10_train_list.append(other_10_train)

    concrete_20_train_list.append(concrete_20_train)
    healthy_20_train_list.append(healthy_20_train)
    incomplete_20_train_list.append(incomplete_20_train)
    irregular_20_train_list.append(irregular_20_train)
    other_20_train_list.append(other_20_train)

    concrete_50_train_list.append(concrete_50_train)
    healthy_50_train_list.append(healthy_50_train)
    incomplete_50_train_list.append(incomplete_50_train)
    irregular_50_train_list.append(irregular_50_train)
    other_50_train_list.append(other_50_train)

  id_series = train_gpd_clean.id
  train_pd = pd.DataFrame({'id': id_series})
  train_pd = pd.concat([train_pd, 
                        pd.DataFrame({'distance_10_all': distance_10_all_list, 'distance_20_all': distance_20_all_list, 'distance_50_all': distance_50_all_list,
                                      'distance_10_train': distance_10_train_list, 'distance_20_train': distance_20_train_list, 'distance_50_train': distance_50_train_list,
                                      'concrete_10_train': concrete_10_train_list, 'healthy_10_train': healthy_10_train_list, 'incomplete_10_train': incomplete_10_train_list, 'irregular_10_train': irregular_10_train_list, 'other_10_train': other_10_train_list,
                                      'concrete_20_train': concrete_20_train_list, 'healthy_20_train': healthy_20_train_list, 'incomplete_20_train': incomplete_20_train_list, 'irregular_20_train': irregular_20_train_list, 'other_20_train': other_20_train_list,
                                      'concrete_50_train': concrete_50_train_list, 'healthy_50_train': healthy_50_train_list, 'incomplete_50_train': incomplete_50_train_list, 'irregular_50_train': irregular_50_train_list, 'other_50_train': other_50_train_list
                                      })], 
                       axis = 1)

  # For test data

  distance_10_all_list, distance_20_all_list, distance_50_all_list = [], [], []
  distance_10_train_list, distance_20_train_list, distance_50_train_list = [], [], []
  concrete_10_train_list, healthy_10_train_list, incomplete_10_train_list, irregular_10_train_list, other_10_train_list = [], [], [], [], []
  concrete_20_train_list, healthy_20_train_list, incomplete_20_train_list, irregular_20_train_list, other_20_train_list = [], [], [], [], []
  concrete_50_train_list, healthy_50_train_list, incomplete_50_train_list, irregular_50_train_list, other_50_train_list = [], [], [], [], []

  for i in range(len(test_gpd_clean)):

    geod = pyproj.Geod(ellps='WGS84')
    def calculate_distance(x, y):
      return geod.inv(test_gpd_clean_coords.iloc[i,0], test_gpd_clean_coords.iloc[i,1], x, y)[2]
    vec_calculate_distance = np.vectorize(calculate_distance)
    distance_from_other_building_train = vec_calculate_distance(train_gpd_clean_coords.centroid_x, train_gpd_clean_coords.centroid_y)
    distance_from_other_building_all = vec_calculate_distance(centroid_coords.centroid_x, centroid_coords.centroid_y)
    
    distance_10_all = np.sum((distance_from_other_building_all < 10) & (distance_from_other_building_all > 0))
    distance_20_all = np.sum((distance_from_other_building_all < 20) & (distance_from_other_building_all > 0))
    distance_50_all = np.sum((distance_from_other_building_all < 50) & (distance_from_other_building_all > 0))

    distance_10_train = np.sum((distance_from_other_building_train < 10) & (distance_from_other_building_train > 0))
    distance_20_train = np.sum((distance_from_other_building_train < 20) & (distance_from_other_building_train > 0))
    distance_50_train = np.sum((distance_from_other_building_train < 50) & (distance_from_other_building_train > 0))

    concrete_10_train   = (np.sum(label_series[(distance_from_other_building_train < 10) & (distance_from_other_building_train > 0)] == 'concrete_cement') + 1) / (distance_10_train + 5)
    healthy_10_train    = (np.sum(label_series[(distance_from_other_building_train < 10) & (distance_from_other_building_train > 0)] == 'healthy_metal') + 1) / (distance_10_train + 5)
    incomplete_10_train = (np.sum(label_series[(distance_from_other_building_train < 10) & (distance_from_other_building_train > 0)] == 'incomplete') + 1) / (distance_10_train + 5)
    irregular_10_train  = (np.sum(label_series[(distance_from_other_building_train < 10) & (distance_from_other_building_train > 0)] == 'irregular_metal') + 1) / (distance_10_train + 5)
    other_10_train      = (np.sum(label_series[(distance_from_other_building_train < 10) & (distance_from_other_building_train > 0)] == 'other') + 1) / (distance_10_train + 5)

    concrete_20_train   = (np.sum(label_series[(distance_from_other_building_train < 20) & (distance_from_other_building_train > 0)] == 'concrete_cement') + 1) / (distance_20_train + 5)
    healthy_20_train    = (np.sum(label_series[(distance_from_other_building_train < 20) & (distance_from_other_building_train > 0)] == 'healthy_metal') + 1) / (distance_20_train + 5)
    incomplete_20_train = (np.sum(label_series[(distance_from_other_building_train < 20) & (distance_from_other_building_train > 0)] == 'incomplete') + 1) / (distance_20_train + 5)
    irregular_20_train  = (np.sum(label_series[(distance_from_other_building_train < 20) & (distance_from_other_building_train > 0)] == 'irregular_metal') + 1) / (distance_20_train + 5)
    other_20_train      = (np.sum(label_series[(distance_from_other_building_train < 20) & (distance_from_other_building_train > 0)] == 'other') + 1) / (distance_20_train + 5)

    concrete_50_train   = (np.sum(label_series[(distance_from_other_building_train < 50) & (distance_from_other_building_train > 0)] == 'concrete_cement') + 1) / (distance_50_train + 5)
    healthy_50_train    = (np.sum(label_series[(distance_from_other_building_train < 50) & (distance_from_other_building_train > 0)] == 'healthy_metal') + 1) / (distance_50_train + 5)
    incomplete_50_train = (np.sum(label_series[(distance_from_other_building_train < 50) & (distance_from_other_building_train > 0)] == 'incomplete') + 1) / (distance_50_train + 5)
    irregular_50_train  = (np.sum(label_series[(distance_from_other_building_train < 50) & (distance_from_other_building_train > 0)] == 'irregular_metal') + 1) / (distance_50_train + 5)
    other_50_train      = (np.sum(label_series[(distance_from_other_building_train < 50) & (distance_from_other_building_train > 0)] == 'other') + 1) / (distance_50_train + 5)

    distance_10_all_list.append(distance_10_all)
    distance_20_all_list.append(distance_20_all)
    distance_50_all_list.append(distance_50_all)
    distance_10_train_list.append(distance_10_train)
    distance_20_train_list.append(distance_20_train)
    distance_50_train_list.append(distance_50_train)

    concrete_10_train_list.append(concrete_10_train)
    healthy_10_train_list.append(healthy_10_train)
    incomplete_10_train_list.append(incomplete_10_train)
    irregular_10_train_list.append(irregular_10_train)
    other_10_train_list.append(other_10_train)

    concrete_20_train_list.append(concrete_20_train)
    healthy_20_train_list.append(healthy_20_train)
    incomplete_20_train_list.append(incomplete_20_train)
    irregular_20_train_list.append(irregular_20_train)
    other_20_train_list.append(other_20_train)

    concrete_50_train_list.append(concrete_50_train)
    healthy_50_train_list.append(healthy_50_train)
    incomplete_50_train_list.append(incomplete_50_train)
    irregular_50_train_list.append(irregular_50_train)
    other_50_train_list.append(other_50_train)

  id_series = test_gpd_clean.id
  test_pd = pd.DataFrame({'id': id_series})
  test_pd = pd.concat([test_pd, 
                        pd.DataFrame({'distance_10_all': distance_10_all_list, 'distance_20_all': distance_20_all_list, 'distance_50_all': distance_50_all_list,
                                      'distance_10_train': distance_10_train_list, 'distance_20_train': distance_20_train_list, 'distance_50_train': distance_50_train_list,
                                      'concrete_10_train': concrete_10_train_list, 'healthy_10_train': healthy_10_train_list, 'incomplete_10_train': incomplete_10_train_list, 'irregular_10_train': irregular_10_train_list, 'other_10_train': other_10_train_list,
                                      'concrete_20_train': concrete_20_train_list, 'healthy_20_train': healthy_20_train_list, 'incomplete_20_train': incomplete_20_train_list, 'irregular_20_train': irregular_20_train_list, 'other_20_train': other_20_train_list,
                                      'concrete_50_train': concrete_50_train_list, 'healthy_50_train': healthy_50_train_list, 'incomplete_50_train': incomplete_50_train_list, 'irregular_50_train': irregular_50_train_list, 'other_50_train': other_50_train_list
                                      })], 
                       axis = 1)

  return train_pd, test_pd

```
